# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
x = np.arange(-10,10,0.01)
def reLU(x):
    return list(map(lambda x:x if x > 0 else 0, x))

plt.plot(x, reLU(x))
plt.grid(color='b',linewidth='0.2' ,linestyle='--')
plt.title('ReLU Function')
```

```python
def dReLU(x):
    return list(map(lambda x : 1 if x > 0 else 0, x))

plt.plot(x, dReLU(x))
plt.grid(color='b',linewidth='0.2' ,linestyle='--')
plt.title('Derivative ReLU Function')
```
