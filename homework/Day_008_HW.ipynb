```python
hw = app_train['AMT_INCOME_TOTAL']
print(hw.describe(),
     type(hw),
     hw.head())

plt.boxplot(np.log(hw))
plt.title('AMT_INCOME_TOTAL')
plt.show()
plt.clf()
plt.hist(hw[hw< 2.025000e+05], bins=200)
plt.title('AMT_INCOME_TOTAL')
plt.show()
plt.clf()
x = np.cumsum(hw[hw< 2.025000e+05].sort_values())
x = x/x.max()
plt.plot(list(x))
plt.grid(True)
plt.show()
```
